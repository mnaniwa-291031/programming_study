# シグモイド関数

ロジスティック回帰で用いられるシグモイド関数（sigmoid function）を、描画しておきます。

```python
%matplotlib inline
import matplotlib.pyplot as pl
import math
import numpy

def sigmoid(x):
  res = 1 / (1 + math.exp(-x))
  return(res)

x = numpy.linspace(-7, 7, 101)
y = [sigmoid(a) for a in x]

pl.figure(figsize=(12, 6))
pl.plot(x, y)
pl.show()
```
